### Ejemplos de ciclo while

```python
i = 1                   # Asigna el valor inicial 1 a la variable i
while i <= 5:           # Mientras i sea menor o igual a 5 (relacional) [9]
    print(i)            # Muestra el valor actual de i
    i += 1              # Incrementa i en 1 (aritmético) para evitar un ciclo infinito [9]
```

```python
numero = 5              # Número del cual se calculará el factorial
factorial = 1           # Acumulador inicializado en 1
while numero > 0:       # Mientras el número sea mayor que 0 (relacional) [11]
    factorial *= numero # Multiplica el acumulador por el número actual (aritmético)
    numero -= 1         # Decrementa el número en 1 (aritmético) [11]
print(factorial)        # Imprime el resultado final (120)
```

```python
suma = 0                # Inicializa el totalizador
while True:             # Ciclo infinito controlado internamente (lógico) [11]
    n = int(input("Ingresa un número: ")) # Solicita y convierte la entrada
    if n < 0:           # Si el número es negativo (relacional)
        break           # Rompe el ciclo inmediatamente (sentencia de salto) [14]
    suma += n           # Suma el número al total (aritmético) [11]
print(f"Total: {suma}") # Muestra la suma acumulada
```

```python
valor = 51              # Comienza la búsqueda desde 51
while valor % 7 != 0:   # Mientras el resto de dividir entre 7 sea distinto de 0 (aritmético y relacional)
    valor += 1          # Incrementa el valor en 1 en cada paso (aritmético)
print(valor)            # Resultado esperado: 56
```

```python
a, b = 0, 1             # Inicializa los dos primeros términos de la serie [15]
while b < 20 and not b > 15: # Mientras b < 20 Y NO sea mayor a 15 (relacional y lógico) [8, 17]
    print(b)            # Imprime el término actual
    a, b = b, a + b     # Actualiza términos sumando los dos anteriores (aritmético) [15]
```
